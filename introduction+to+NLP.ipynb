```python
import pandas as pd
submissions = pd.read_csv("F:/education/data quest/machine learning sessions/introduction to NLP/sel_hn_stories.csv")
submissions.columns = ["submission_time", "upvotes", "url", "headline"]
submissions = submissions.dropna()
```

```python
tokenized_headlines = []
for item in submissions["headline"]:
    l = list(item.strip().split(" "))
    tokenized_headlines.append(l)
print(tokenized_headlines[1:5])
```

```python
punctuation = [",", ":", ";", ".", "'", '"', "’", "?", "/", "-", "+", "&", "(", ")"]
clean_tokenized = []
for item in tokenized_headlines:
    tokens = []
    for token in item:
        token = token.lower()
        for punc in punctuation:
            token = token.replace(punc, "")
        tokens.append(token)
    clean_tokenized.append(tokens)
```
